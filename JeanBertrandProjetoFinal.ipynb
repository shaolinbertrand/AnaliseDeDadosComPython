### 1) Como foi a evolução das exportações para a França ao longo dos anos?

Dica: Crie uma tabela com os anos de 2016, 2017, 2018, 2019 e 2020 e o valor US$ de exportações

```python
#filtrar anos desejados
anos_desejados = [2016,2017,2018,2019,2020]
df_franca_filtrado = tabela[tabela['Year'].isin(anos_desejados)]

# 4. Agrupar por 'Year' e somar os valores de 'US$ FOB'
evolucao_exportacoes = df_franca_filtrado.groupby('Year')['US$ FOB'].sum().reset_index()

# 5. Criar a tabela final com os anos e os valores de US$ de exportações
# Renomear a coluna para algo mais descritivo, se desejar
evolucao_exportacoes.rename(columns={'US$ FOB': 'Total US$ FOB Exportado'}, inplace=True)

# Exibir a tabela resultante
print("Evolução das Exportações para a França (2016-2020):")
print(evolucao_exportacoes)
```

### 2) Quais os produtos mais exportados ao longo de todo o período?

Queremos saber os produtos com maior valor US$ de exportação

```python
# 2. Agrupar por 'SH4 Description' e somar os valores de 'US$ FOB'
# Isso vai totalizar o valor de exportação para cada tipo de produto.
produtos_mais_exportados = tabela.groupby('SH4 Description')['US$ FOB'].sum().reset_index()

# 3. Classificar os produtos em ordem decrescente de valor de exportação
# Assim, os produtos com maior valor aparecerão no topo.
produtos_mais_exportados = produtos_mais_exportados.sort_values(by='US$ FOB', ascending=False)

# 4. Exibir a tabela resultante
print("Top 10 Produtos Mais Exportados (em US$ FOB) ao longo de todo o período:")
print(produtos_mais_exportados.head(10))
```

### 3) Em 2020 qual cidade mais exportou para a França?
Em valores US$

```python
df_franca_2020 = df_franca_filtrado[df_franca_filtrado['Year'] == 2020] #Pegar apenas o ano de 2020
df_franca_2020
cidades_que_mais_exportaram = df_franca_2020.groupby('City')['US$ FOB'].sum().reset_index() #agrupando pela cidade que mais faturou
cidades_que_mais_exportaram = cidades_que_mais_exportaram.sort_values(by='US$ FOB',ascending=False) #ordenando de forma decrescente
#Exibindo a tabela com as cidades que mais exportaram em 2020
print("10 Cidades que mais exportaram para França em 2020")
print(cidades_que_mais_exportaram.head(10))
```

### 4) Quais os produtos mais exportados (em US$) que as 2 maiores cidades (em exportação em 2020) exportaram?

```python
cidades_top2 = cidades_que_mais_exportaram.head(2)
nomes_top2 = cidades_top2['City'].tolist() #pegando apenas o nome das duas cidades que mais exportaram em 2020
produtos_mais_exportados_cidades_top2 = df_franca_2020[df_franca_2020['City'].isin(nomes_top2)]
produtos_mais_exportados_cidades_top2 = produtos_mais_exportados_cidades_top2.groupby('SH4 Description')['US$ FOB'].sum().reset_index()
produtos_mais_exportados_cidades_top2 = produtos_mais_exportados_cidades_top2.sort_values(by='US$ FOB',ascending=False)
print("Top 10 produtos mais exportados pelas duas cidades que mais exportaram em 2020")
print(produtos_mais_exportados_cidades_top2.head(10))
```
